```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data/raw_data.csv")

df.head()

```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["productivity_score"], bins=20, kde=True)
plt.title("Distribution of Productivity Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="sleep_hours",y="productivity_score",data=df,alpha=0.5
)
plt.title("Sleep Hours vs Productivity")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="focus_level",y="productivity_score",data=df
)
plt.title("Focus Level vs Productivity")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="task_type",y="productivity_score",data=df)
plt.title("Productivity by Task Type")
plt.show()
```

```python
coding_df = df[df["task_type"] == "coding"]

plt.figure(figsize=(6,4))
sns.boxplot(
    x="music_type",
    y="productivity_score",
    data=coding_df
)
plt.title("Music Type Impact on Coding Productivity")
plt.show()
```

### Key EDA Insights

- Productivity shows a realistic spread between 0 and 100.
- Low sleep hours strongly cap productivity, even for high focus levels.
- Higher focus generally leads to higher productivity, but does not guarantee it.
- Coding tasks are more sensitive to mood and music choice than routine tasks.
- Instrumental and lofi music support coding productivity better than lyrical music.


### Music Type vs Coding Productivity

- Lofi music shows the highest median productivity for coding tasks, indicating stable and sustained focus.
- Instrumental music allows peak productivity in some sessions but shows higher variability.
- Silence provides moderate productivity and appears safer than lyrical music.
- Lyrical music consistently results in lower productivity, likely due to cognitive distraction.
- These patterns align with the rule-based assumptions used during data generation.
